```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
path = 'C:/Users/lyc07/Desktop/Columbia/S3/GR5243 Applied Data Science/prj3/'
dat_train = pd.read_csv(path + "dat_train.csv")
dat_test = pd.read_csv(path + 'dat_test.csv')
```

```python
dat_train.head()
```

```python
dat_full = pd.concat([dat_train, dat_test], ignore_index = True)
```

```python
dat_full.head()
```

```python
dat_full.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
X, y = dat_full.iloc[:,:-1],dat_full.iloc[:,-1]
```

```python
dat_full_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
# gamma
xg_cls = xgb.XGBClassifier(objective ='multi:softmax', colsample_bytree = 0.3, learning_rate = 0.3,
                max_depth = 3, num_class = 22, n_estimators = 10, num_features = 10)
```

```python
xg_cls.fit(X_train,y_train)
```

```python
preds = xg_cls.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, preds)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
