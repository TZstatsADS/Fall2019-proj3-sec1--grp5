```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
path = '//Users/tingcai/Desktop/Current_Curriculum/STATGR5243/Project3-FacialEmotionRecognition/doc/'
dat_train = pd.read_csv(path + "dat_train.csv")
dat_test = pd.read_csv(path + 'dat_test.csv')
```

```python
dat_train
```

```python
dat_full = pd.concat([dat_train, dat_test], ignore_index = True)
```

```python
dat_full.head()
```

```python
X, y = dat_full.iloc[:,:-1],dat_full.iloc[:,-1]
```

#### Splite into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=16)
```

#### Use 20% of the train to tune parameter

```python
X_train_s, X_tune, y_train_s, y_tune = train_test_split(X_train, y_train, test_size = 0.2, random_state=3)
```

#### SVM-linear

```python
param_grid = {'C': [0.1, 0.01, 0.0001, 0.00001, 0.000001]}
```

```python
clf_grid = GridSearchCV(svm.SVC(kernel='linear'), param_grid, verbose=1)
 
# Train the classifier
clf_grid.fit(X_tune, y_tune)
 
# clf = grid.best_estimator_()
print("Best Parameters:\n", clf_grid.best_params_)
print("Best Estimators:\n", clf_grid.best_estimator_)
```

```python
clf_best = svm.SVC(kernel='linear', C = 0.1)
clf_best.fit(X_train, y_train)
clf_best_predictions = clf_best.predict(X_test)
print("Accuracy: {}%".format(clf_best.score(X_test, y_test) * 100 ))
```

```python
clf = svm.SVC(kernel='linear', C = 0.00001)
clf.fit(X_train, y_train)
clf_predictions = clf.predict(X_test)
print("Accuracy: {}%".format(clf.score(X_test, y_test) * 100 ))
```
